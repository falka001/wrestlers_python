```python
import numpy as np
import pandas as pd
import string
import random as rd
```

```python
def wlr_grps(grprect):
    # assigning wrestlers to groups where grprect is the list of wrestlers drafted
    grpA = []
    grpB = []
    grpC = []
    grpD = []
    grpE = []
    grpF = []
    grpG = []
    grpH = []
    grpI = []
    grpJ = []
    print("drafted wrestlers are being placed in group according to their weights")
    for i in grprect:
        if i > 87 and i <= 106:
            grpA.append(i)
            grprect.remove(i)
        if i > 106 and i <= 113:
            grpB.append(i)
            grprect.remove(i)
        if i > 114 and i <= 120:
            grpC.append(i)
            grprect.remove(i)
        if i > 121 and i <= 126:
            grpD.append(i)
            grprect.remove(i)
        if i > 127 and i <= 132:
            grpE.append(i)
            grprect.remove(i)
        if i > 133 and i <= 138:
            grpF.append(i)
            grprect.remove(i)
        if i > 139 and i <= 145:
            grpG.append(i)
            grprect.remove(i)
        if i > 146 and i <= 152:
            grpH.append(i)
            grprect.remove(i)
        if i > 153 and i <= 160:
            grpI.append(i)
            grprect.remove(i)
        if i > 161 and i <= 170:
            grpJ.append(i)
            grprect.remove(i)
    print("Wrestlers not qualified:")
    print("  " + str(grprect))
    print(str(grpA) + "\n" + str(grpB) + "\n" + str(grpC) + "\n" + str(grpD) + "\n" + str(grpE) + "\n" + str(grpF) + "\n" + str(grpG) + "\n" + str(grpH) + "\n" + str(grpI)
      + "\n" + str(grpJ))
    Groups = [grpA, grpB, grpC, grpD, grpE, grpF, grpG, grpH, grpI, grpJ]
    print("\nGroups:\n" + str(Groups))
    return Groups
```

```python
def scores(teama, teamb):
    ab_scA = []
    ab_scB = []
    ab_scC = []
    for i in range(len(teama)):
        for j in range(len(teama[i])):
            weight = teama[i][j]
            rd_num = int(rd.triangular(10, 100, 55))
            ability = weight / rd_num
            ab_scA.append(int(ability))
    for i in range(len(teamb)):
        if len(teamb[i]) == 0:
            ab_scB.append(0)
        for j in range(len(teamb[i])):
            weight = teamb[i][j]
            rd_num = int(rd.triangular(10, 100, 55))
            ability = weight / rd_num
            ab_scB.append(int(ability))
    print("Team A ability scores:\n" + str(ab_scA))
    print("\n       Number of wrestlers = " + str(len(ab_scA)))
    print("\nTeam B ability scores:\n" + str(ab_scB))
    print("\n       Number of wrestlers = " + str(len(ab_scB)))
    ab_sc = [ab_scA, ab_scB]
    return ab_sc
```

```python
lst1 = []
j = 1
for i in string.ascii_uppercase:
    if j <= 10:
        lst1.append(i)
    if j == 11:
        break
    j = j+1

wrestlersA = []
wrestlersB = []
for i in range(0,201):
    if i < 100:
        wrestlersA.append(int(rd.triangular(50, 180, 65)))
    if i > 100:
        wrestlersB.append(int(rd.triangular(50, 180, 65)))
print("Wrestlers drafted for Team A:\n")
print(str(wrestlersA[0:5]) + "...  end of list  ..." + str(wrestlersA[95:100]) + "\n          " + str(len(wrestlersA)) + " wrestlers have been drafted")
print("\nWrestlers drafted for Team B:\n")
print(str(wrestlersB[0:5]) + "...  end of list  ..." + str(wrestlersB[95:100]) + "\n          " + str(len(wrestlersA)) + " wrestlers have been drafted")

print("\n")

print("Team A:\n")
TA = wlr_grps(wrestlersA)
print("\nTeam B:\n")
TB = wlr_grps(wrestlersB)

ab_scAB = scores(TA, TB)

TA_scores = ab_scAB[0]
TB_scores = ab_scAB[1]

print("\n")

data = {'Team 1': [len(TA[0]), len(TA[1]), len(TA[2]), len(TA[3]), len(TA[4]), len(TA[5]), len(TA[6]), len(TA[7]), len(TA[8]), len(TA[9])]}
match = pd.DataFrame(data, index = lst1 , columns = ['Team 1'])
match['Team 2'] = [len(TB[0]), len(TB[1]), len(TB[2]), len(TB[3]), len(TB[4]), len(TB[5]), len(TB[6]), len(TB[7]), len(TB[8]), len(TB[9])]
print(match)

suma = 0
sumb = 0
sumA = []
sumB = []
winner = []
for i in range(len(TA)):
    for j in range(len(TA[i])):
        suma += TA_scores[j]
    sumA.append(suma)
    suma = 0
for i in range(len(TB)):
    for j in range(len(TB[i])):
        sumb += TB_scores[j]
    sumB.append(sumb)
    sumb = 0
for i in range(0,10):
    if sumA[i] < sumB[i]:
        winner.append('B')
    if sumA[i] > sumB[i]:
        winner.append('A')
    if sumA[i] == sumB[i]:
        winner.append('no winner')
print("\nTeam A Scores Per Group:\n")
print(sumA)
print("\nTeam B Scores Per Group:\n")
print(sumB)
print("\n" + str(winner))
match['A_scores'] = sumA
match['B_scores'] = sumB
match['WINNER'] = winner
print("\nSummary:\n")
match
```
